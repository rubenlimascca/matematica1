```python
import numpy.polynomial as P
p=P.Polynomial([1,2,3])
print(p)
r=P.Polynomial([-1,0,0,1])
print(r)
print(r.degree())
print(r.coef)
print(r.roots()) # Permite calcular las raices del polinomio
print(r.coef[0]) # Extrae el primer coeficiente
q=P.Polynomial.fromroots([-1,1]) # nos da el polinomio monico (coeficientre 1) de menor grado 
print(q)
```
